```python
import pandas as pd
import numpy as np

df = terror_events.merge(location,left_on="location_id", right_on= "id")

def analyze_attack_locations(df):
    location_analysis = (df.groupby(['country', 'city', 'longitude', 'latitude'])['terror_group']
                        .nunique()
                        .reset_index()
                        .rename(columns={'terror_group': 'groups_count'}))
    
    # מציאת המיקום עם מספר הקבוצות הגדול ביותר עבור כל מדינה
    results = (location_analysis.sort_values('groups_count', ascending=False)
               .groupby('country')
               .first()
               .reset_index())
    # 
    def get_terror_groups(row):
        mask = ((df['longitude'] == row['longitude']) & 
                (df['latitude'] == row['latitude']))
        return df[mask]['terror_group'].unique()
    
    results['terror_groups'] = results.apply(get_terror_groups, axis=1)
    
    return results
    




res = analyze_attack_locations(df)
res
```
